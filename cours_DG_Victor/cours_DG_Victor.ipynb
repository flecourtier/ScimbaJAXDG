## 1. Équation de transport linéaire

$$ E(u) = \frac{\partial u}{\partial t} + \frac{\partial f(u)}{\partial x} = 0 \qquad u : [0, T] \times \Omega \rightarrow \mathbb{R} $$

$$ f \in \mathcal{C}^1(\mathbb{R}, \mathbb{R}) $$

**Exemple :** $$ f(u) = \alpha u $$

## 2. Espaces fonctionnels et Maillage

$u \in V$ (espace fonctionnel de dim infinie)

**Exemple :** $V = L^2(\mathbb{R})$

$V_{h,i} \subset V$ (espace fonctionnel, ss esp vect de $V$ de dim finie : $p+1$) $\leftarrow$ dépend du maillage et de la cellule $i$

$$ V_{h,i} = \text{Span}(\varphi_i^0, \dots, \varphi_i^p) $$
où $i$ est le numéro de cellule du maillage $\mathcal{M}$.

### Maillage 1D
$$ i \in \mathcal{M} \subset \mathbb{Z} $$
$$ x_i = a + ih + \frac{h}{2} $$
$$ x_{i+\frac{1}{2}} = a + \left(i+\frac{1}{2}\right)h + \frac{h}{2} $$

## 3. Bases polynomiales locales

$\forall i \in \mathcal{M}, \forall x \in \left(x_{i-\frac{1}{2}}, x_{i+\frac{1}{2}}\right) = C_i$

**Exemple :** Base de Taylor
$$ \varphi_i^{(k)}(x) = \frac{(x-x_i)^k}{k!} \qquad \forall k \in [\![0, p]\!] \qquad (\circledast)$$

**Exemple :** Base de Lagrange

$(x_l^G)_{l \in [\![0, p]\!]}$ points de quadrature (pt de Gauss)

$$ \varphi_i^{(k)}(x) = \frac{\prod_{\substack{l=0 \\ l \neq k}}^p (x - x_l^G)}{\prod_{\substack{l=0 \\ l \neq k}}^p (x_k^G - x_l^G)} \qquad \forall k \in [\![0, p]\!] $$

## 4. Développement de la solution

$(\circledast) \quad \forall i \in \mathcal{M}, \forall x \in C_i, \quad u_{h,i}(x) = \sum_{j=0}^p u_{i,j} \varphi_i^{(j)}(x)$

**Exemple :** $p=3$
$$ u_{h,i}(x) = u_{i,0} + u_{i,1}(x-x_i) + u_{i,2}\frac{(x-x_i)^2}{2} + u_{i,3}\frac{(x-x_i)^3}{6} $$

$\longrightarrow$ ($p+1$) degrés de liberté (ddl) par cellule.

## 5. Solution globale et Projection

$$ \forall x \in \Omega \qquad u_h(x) = \sum_{i \in \mathcal{M}} u_{h,i}(x) \mathbb{1}_{\{x \in C_i\}} $$

Soit $i \in \mathcal{M}$, on cherche $u_{h,i}$ comme la fonction de $V_{h,i}$ qui réalise la projection orthogonale de $E(u_{h,i})$ sur $V_{h,i}$ :

$$ \langle E(u_{h,i}) , \varphi_i^{(k)} \rangle_{L^2} = 0 \qquad \forall k \in [\![0, p]\!] $$

## 6. Formulation variationnelle locale

$$ \forall i \in \mathcal{M}, \forall k \in [\![0, p]\!], \forall t \in [t^n, t^n + \Delta t] $$

$$ \int_{C_i} E(u_{h,i})(t,x) \varphi_i^{(k)}(x) dx = 0 \qquad (\circledast) $$

On remplace par le développement de $u_{h,i}$ :

$$ u_{h,i}(t,x) = \sum_{j=0}^p u_{i,j}(t) \varphi_i^{(j)}(x) $$

$$ E(u_{h,i})(t,x) = \sum_{j=0}^p \frac{\partial}{\partial t} u_{i,j}(t) \varphi_i^{(j)}(x) + \alpha \sum_{j=0}^p u_{i,j}(t) \frac{\partial}{\partial x} \varphi_i^{(j)}(x) $$
*(pour le cas spécifique où $f(u) = \alpha u$)*

> **Note en marge :** On discrétise en temps et on fait Euler explicite.

## 7. Développement de l'intégrale et Discrétisation

$\forall i \in \mathcal{M}, \forall k \in [\![0, p]\!], \forall t \in [t^n, t^n + \Delta t]$,

L'équation $(\circledast)$ devient :

$$ \underbrace{\sum_{j=0}^p \frac{d}{dt} u_{i,j}(t) \int_{C_i} \varphi_i^{(j)}(x) \varphi_i^{(k)}(x) dx}_{\textcolor{red}{\text{Terme \textcircled{1}}}} + \underbrace{\int_{C_i} \frac{\partial}{\partial x} f(u_{h,i}(t,x)) \varphi_i^{(k)}(x) dx}_{(**)} = 0 $$

## 8. Intégration par parties et Flux aux interfaces

On reprend le terme $(**)$ de l'équation précédente et on intègre par parties :

$$ (**) = \underbrace{\left[ f(u_{h,i}(t,x)) \varphi_i^{(k)}(x) \right]_{x_{i-\frac{1}{2}}}^{x_{i+\frac{1}{2}}}}_{(**\ast)} - \underbrace{\int_{C_i} f(u_{h,i}(t,x)) \frac{d}{dx} \varphi_i^{(k)}(x) dx}_{\textcolor{red}{\text{Terme \textcircled{2}}}} $$

On développe le terme de bord $(***)$ :
$$ (**\ast) = \underbrace{f(u_{h,i}(t, x_{i+\frac{1}{2}})) \varphi_i^{(k)}(x_{i+\frac{1}{2}}) - f(u_{h,i}(t, x_{i-\frac{1}{2}})) \varphi_i^{(k)}(x_{i-\frac{1}{2}})}_{\textcolor{red}{\text{Terme \textcircled{3}}}} $$

## 9. Flux Numérique

Aux interfaces, la solution est discontinue, on approxime donc le flux par un **flux numérique** $\mathcal{F}$ :

$$ f(u_{h,i}(t, x_{i+\frac{1}{2}})) \simeq \mathcal{F}(u_{h,i}(t, x_{i+\frac{1}{2}}), u_{h,i+1}(t, x_{i+\frac{1}{2}})) $$

$\mathcal{F}$ est un flux numérique consistant, tel que $\mathcal{F}(v,v) = f(v) \quad \forall v \in \mathbb{R}$.

**Exemples :**

* *Flux centré :* 
$$\mathcal{F}(v_L, v_R) = \frac{1}{2}(f(v_L) + f(v_R))$$
* *Flux décentré :* 
$$\mathcal{F}(v_L, v_R) = \frac{1}{2}(f(v_L) + f(v_R)) - D(v_L, v_R)(v_R - v_L)$$

## 10. Discrétisation temporelle (Euler explicite) et Schéma Numérique Final

* **V1 :** On le fait avec une intégrale.
* **V2 :** On approxime la dérivée temporelle et on évalue au temps $t^n$ :

$$ \frac{\partial}{\partial t} u_{i,j}(t) \simeq \frac{u_{i,j}(t^{n+1}) - u_{i,j}(t^n)}{\Delta t} \quad \text{et} \quad u_{i,j}(t) \simeq u_{i,j}(t^n) $$

On note $u_{i,j}(t^n) = u_{i,j}^n$.

En assemblant tous les termes (1, 2 et 3), on obtient le schéma totalement discret :

$$ \sum_{j=0}^p \frac{u_{i,j}^{n+1} - u_{i,j}^n}{\Delta t} \underbrace{\int_{C_i} \varphi_i^{(j)}(x) \varphi_i^{(k)}(x) dx}_{M_i = (M_{j,k})_{j,k \in [\![0,p]\!]} \; \textcolor{red}{\textcircled{1}}} - \underbrace{\int_{C_i} f\left(\sum_{j=0}^p u_{i,j}^n \varphi_i^{(j)}(x)\right) \frac{d}{dx} \varphi_i^{(k)}(x) dx}_{V_i(U_i^n) \; \textcolor{red}{\textcircled{2}}} $$

$$ + \underbrace{\left\{ \mathcal{F}\left(\sum_{j=0}^p u_{i,j}^n \varphi_i^{(j)}(x_{i+\frac{1}{2}}), \sum_{j=0}^p u_{i+1,j}^n \varphi_{i+1}^{(j)}(x_{i+\frac{1}{2}})\right) - \mathcal{F}\left(\sum_{j=0}^p u_{i-1,j}^n \varphi_{i-1}^{(j)}(x_{i-\frac{1}{2}}), \sum_{j=0}^p u_{i,j}^n \varphi_i^{(j)}(x_{i-\frac{1}{2}})\right) \right\}}_{F_i(U_{i-1}^n, U_i^n, U_{i+1}^n) \; \textcolor{red}{\textcircled{3}}} = 0 $$
